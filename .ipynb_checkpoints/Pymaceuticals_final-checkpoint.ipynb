### Analysis

- The overall analysis is that compared with the other drugs, Capomulin was most effective at reducing tumor volume.
- The two drugs which showed the most reduction in tumor growth were Capomulin and Ramicane, over 45 days, the tumor size reduced as shown through the box plot diagram as well as the Summary data frame.
- Capomulin and Ramicane have the lowest average tumor volume (mm3) as seen from the Summary data frame as well as through the box plot diagram. It's interesting to note that these drugs also had the highest number of timepoints for the datasets.
- From observig the effectiveness of the drug from a daily standpoint, the line chart shows the pattern in which Capomulin reduces the tumor volume, spiking twice before reducing the size significantly.



 

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np

# Study data files
study_results_path = 'Study_results.csv'
mouse_metadata_path = 'Mouse_metadata.csv'


# Read the mouse data and the study results
study_results = pd.read_csv(study_results_path)
mouse_metadata = pd.read_csv(mouse_metadata_path)


# Combine the data into a single DataFrame
combined_df = pd.merge(mouse_metadata, study_results, how='right', on="Mouse ID")

# Display the data table for preview
combined_df
```

```python
# Checking the number of mice.
number_mice = combined_df["Mouse ID"].count()
number_mice
```

```python
# Our data should be uniquely identified by Mouse ID and Timepoint
# Get the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 
dupe_miceid = combined_df[combined_df.duplicated(["Mouse ID", "Timepoint"])]
dupe_miceid
```

```python
# Optional: Get all the data for the duplicate mouse ID. Mouse ID g989 

combined_df.loc[combined_df["Mouse ID"]=="g989",:]
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
Clean_df = combined_df[combined_df["Mouse ID"] != 'g989']
Clean_df
```

```python
# Checking the number of mice in the clean DataFrame.
new_number_mice = Clean_df["Mouse ID"].unique()
number_mice = len(new_number_mice)
number_mice
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen
# Use groupby and summary statistical methods to calculate the following properties of each drug regimen:
# mean, median, variance, standard deviation, and SEM of the tumor volume. 
mean = Clean_df.groupby("Drug Regimen")["Tumor Volume (mm3)"].mean().map('{:.2f}'.format)
median = Clean_df.groupby("Drug Regimen")["Tumor Volume (mm3)"].median().map('{:.2f}'.format)
var = Clean_df.groupby("Drug Regimen")["Tumor Volume (mm3)"].var().map('{:.2f}'.format)
stddev = Clean_df.groupby("Drug Regimen")["Tumor Volume (mm3)"].std().map('{:.2f}'.format)
sem = Clean_df.groupby("Drug Regimen")["Tumor Volume (mm3)"].sem().map('{:.2f}'.format)
 
# Assemble the resulting series into a single summary DataFrame.
statistics_df = pd.DataFrame({"Mean Tumor Volume": mean, 
                              "Median Tumor Volume": median, 
                              "Tumor Volume Variance": var,
                              "Tumour Volume Std Dev.": stddev,
                              "Tumor Volume Std Err.": sem})
statistics_df
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin


# Start by getting the last (greatest) timepoint for each mouse
last_timepoint = Clean_df.groupby(['Mouse ID'])
max_timepoint_byMouse_ID = last_timepoint['Timepoint'].max()
pd.DataFrame(max_timepoint_byMouse_ID)
max_timepoint_byMouse_ID
# Merge this group df with the original DataFrame to get the tumor volume at the last timepoint
merged_original = pd.merge(Clean_df, max_timepoint_byMouse_ID, how='right', on=["Mouse ID", "Timepoint"])
merged_original
```

```python
# Put treatments into a list for for loop (and later for plot labels)
drug_regimens = ['Capomulin','Ramicane','Infubinol','Ceftamin']

# Create empty list to fill with tumor vol data (for plotting)
tumor_vol = []

# Calculate the IQR and quantitatively determine if there are any potential outliers. 

for drug in drug_regimens:
    
    # Locate the rows which contain mice on each drug and get the tumor volumes
    per_drug_tumor = merged_original.loc[merged_original['Drug Regimen']==drug,'Tumor Volume (mm3)']
    
    # add subset 
    tumor_vol.append(per_drug_tumor)
    
    # Determine outliers using upper and lower bounds
    quartiles = per_drug_tumor.quantile([.25, .5, .75])
    lowerq = quartiles[0.25]
    upperq = quartiles[0.75]
    iqr = upperq-lowerq

    lower_bound = lowerq - (1.5*iqr)
    upper_bound = upperq + (1.5*iqr)
    print(f'The potential outlier for {drug_regimens} is: {iqr}')
    
```

```python
# Generate a box plot that shows the distrubution of the tumor volume for each treatment group.
fig1, ax1 = plt.subplots()

ax1.set_ylabel('Final Tumor Volume (mm3)')
ax1.set_xlabel('Drug Regimen')
ax1.boxplot(tumor_vol)
ax1.set_xticklabels(drug_regimens)
plt.show()
```

```python
# Generate a scatter plot of mouse weight vs. the average observed tumor volume for the entire Capomulin regimen
# Create the lists for mouse weight and ave tumor volume 
capo_mice = Clean_df.loc[Clean_df['Drug Regimen']=='Capomulin']
capo_mice_weight = capo_mice.groupby(['Mouse ID']).mean()['Weight (g)']
capo_mice_tumor_vol = capo_mice.groupby(['Mouse ID']).mean()['Tumor Volume (mm3)']

plt.scatter(capo_mice_weight, capo_mice_tumor_vol, marker='o', facecolors='steelblue')
plt.xlabel('Weight (g)')
plt.ylabel('Average Tumor Volume (mm3)')
plt.show()
```

## Correlation and Regression

```python
# Calculate the correlation coefficient
# for mouse weight and average observed  tumor volume for the entire Capomulin regimen
corr_calc = st.pearsonr(capo_mice_weight, capo_mice_tumor_vol)
print(f"The correlation coefficient between mouse weight and the average tumor volume is {round(corr_calc[0],3)}")
```

```python
# Create a linear regression model for mouse weight and average observed  tumor volume for the entire Capomulin regimen
vc_slope, vc_int, vc_r, vc_p, vc_std_err = st.linregress(capo_mice_weight, capo_mice_tumor_vol)
vc_fit = vc_slope * capo_mice_weight + vc_int

plt.scatter(capo_mice_weight, capo_mice_tumor_vol, marker='o', facecolors='orange')
plt.xlabel('Weight (g)')
plt.ylabel('Average Tumor Volume (mm3)')
plt.plot(capo_mice_weight, vc_fit, "-")
plt.show()
```
